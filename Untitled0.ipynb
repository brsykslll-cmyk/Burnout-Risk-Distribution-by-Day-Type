```python
import pandas as pd
import matplotlib.pyplot as plt


titanic= pd.read_csv("titanic.csv") # Assuming the file is directly uploaded or in the current working directory


titanic.info()

#boş kalan yerler varsa toplar ve bize sayısını bildirir
titanic.isnull().sum()

#bize bu tablodaki indexleri yani kolon isimlerini verir
titanic.columns

#mod medyan avg gibi değerleri renkli bir tabloda gösterir
titanic.describe().style.background_gradient(cmap="Spectral_r")

titanic["Survived"] = titanic["Survived"].map({0: "dead", 1: "survived"})
#titanic.describe().style.background_gradient(cmap="Spectral_r")

# Calculate survival percentages
survival_percentages = titanic["Survived"].value_counts(normalize=True) * 100

# Plotting the bar chart for percentages
survival_percentages.plot(kind="bar")
plt.title("Hayatta Kalma Yüzdesi")
plt.xlabel("Durum")
plt.ylabel("Yüzde (%)")
plt.ylim(0, 100) # Set y-axis limit to 0-100 for percentages
plt.xticks(rotation=0) # Ensure labels are horizontal

# Add percentage labels on top of the bars
for index, value in enumerate(survival_percentages):
    plt.text(index, value + 2, f'{value:.2f}%', ha='center', va='bottom')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Calculate the average screen time for each day type
screen_time_by_day_type = df.groupby('day_type')['screen_time_hours'].mean().reset_index()

# Define the order for day_type for better visualization (optional but good practice)
screen_time_by_day_type['day_type'] = pd.Categorical(screen_time_by_day_type['day_type'], categories=['Weekday', 'Weekend'], ordered=True)
screen_time_by_day_type = screen_time_by_day_type.sort_values('day_type')

# Create the line plot
plt.figure(figsize=(8, 5))
sns.lineplot(x='day_type', y='screen_time_hours', data=screen_time_by_day_type, marker='o')

plt.title('Ortalama Ekran Süresi, Haftanın Gününe Göre')
plt.xlabel('Gün Tipi')
plt.ylabel('Ortalama Ekran Süresi (Saat)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
